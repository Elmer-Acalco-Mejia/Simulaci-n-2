# **5 Escriba un programa en python para realizar ChiCuadrado**


```python
from random  import uniform
import numpy as np
import pandas as pd
```

Solicitamos que se ingrese la cantidad de elementos a las que se someterá la prueba , el número de intervalos y creamos nuestros elementos usando la función "uniform" entre el intervalo(0,1)

```python
n=40
intervalos=5
numeros=[]
d=1/intervalos
for i in range(n):
  z=uniform(0,1)
  numeros.append(z)
```

Creamos una lista que contendrá el número de listas de acuerdo al tamaño del intervalo ; es decir l=[[ ],[ ],...intervalos] con el objetivo de ir agregando los elementos generados en la lista "numeros" de acuerdo al intervalo en el que se encuentren ,por ejemplo l=[[ ],[ ],...] en la primera lista dentro de la lista se almacenará los elementos que se encuentran entre 0-0.2 , en la segunda lista se almacenará los elementos que se encuentran entre 0.2-0.4 , etc.

```python
lista3=[]
for i in range(intervalos):
  lista3.append([])
  for j in numeros:
    if i*d<j<(i+1)*d:
      lista3[i].append(j)
```

Contamos la cantidad de elementos que contiene cada lista de la lista "lista3" , con el propósito de conocer la frecuencia observada en cada intervalo

```python
cantidad=[]
for fila in lista3:
  cantidad.append(len(fila))
print(cantidad)
```

Creamos un diccionario para poder vizualizar mejor la información en una data Frame

```python
esperado=n/intervalos  # Calculamos la Frecuencia Esperada
tabla_analisis={'Frecuencia Observada O_i':cantidad,'Frecuencia Esperada E_i':[esperado for i in range(intervalos)],'O_i -E_i':[i-esperado for i in cantidad],'Chi_cuadrada':[((i-esperado)**2)/esperado for i in cantidad]}
tabla_analisis
Analisis_ChiCuadrada=pd.DataFrame(tabla_analisis)
Analisis_ChiCuadrada
```

Creamos los intervalos en los que se metieron los elementos generados al inicio (n=40) Agregamos los intervalos como índice al data Frame


```python
indice=[]
dt=0
while dt<1:
  ent=str(dt)+'-'+str(round(dt+d,2))
  indice.append(ent)
  dt=round(dt+d,2)
print(indice)
Analisis_ChiCuadrada.index=indice
Analisis_ChiCuadrada
```

Calculamos el valor de la Chi Cuadrada sumando la columna Chi_cuadrada del Data Frame

```python
Valor_ChiCuadrada=Analisis_ChiCuadrada['Chi_cuadrada'].sum()
print('El valor de la Chi Cuadrada es :',Valor_ChiCuadrada)
```

Formulación de la hipótesis nula y alternativa Ho: No hay diferencia entre la distribución de la muestra y la distribución teórica

Si Chi_cuadrada > Valor Crítico entonces se rechaza la hipótesis Ho

Si Chi_cuadrada <= Valor crítico entonces no se rechaza la hipótesis nula Ho

```python
from scipy.stats import chi2
gl=intervalos-1
#Encontramos el valor crítico de chi-cuadrado con un nivel de significancia de 0.05
valor_critico=chi2.ppf (1-0.05,gl )
print("El valor crítico es ",valor_critico)
if Valor_ChiCuadrada<=valor_critico:
    print("No se rechaza la hipótesis nula Ho ")
else:
    print("Rechazar la hipótesis nula Ho ")
```
